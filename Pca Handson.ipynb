```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
x_scaled = scalar.fit_transform(x)
x_scaled
```

```python
x_scaled.T
```

```python
cov_mat=np.cov(x_scaled.T)
pd.DataFrame(cov_mat)
```

```python
eig_val,eig_vec = np.linalg.eig(cov_mat)
pd.DataFrame(eig_val)
```

```python
pd.DataFrame(eig_vec)
```

```python
total = sum(eig_val)
```

```python
exp_var=[]

for i in sorted(eig_val,reverse=True):
    exp_var.append((i/total)*100)

pd.DataFrame(exp_var,columns=['Explained Variance'])
```

```python
cum_exp_var = np.cumsum(exp_var)
pd.DataFrame(cum_exp_var,columns=["Cummulative Frequency EXP-VAR"])
```

```python
plt.figure(figsize =(7,4))
plt.bar(range(len(exp_var)),exp_var,label="Individual explained variance",color ="g")
plt.step(range(len(exp_var)),cum_exp_var,label="Cumulative explained variance")
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance")
plt.axhline(y=1,color = "red",linestyle = "--")
plt.legend()
plt.show()
```
